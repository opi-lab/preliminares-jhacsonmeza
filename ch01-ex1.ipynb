**1.** Take an image and apply Gaussian blur like in Figure 1-9. Plot the image contours for increasing values of σ. What happens? Can you explain why?


Para hacer estas pruebas vamos a hacer uso de la librería *Scipy* que tiene un modulo que nos calcula $I \ast G_{\sigma}$ para el valor de sigma dado. Para realizar las pruebas, usaremos la siguiente imagen:

```python
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

im = np.array(Image.open('data/fresas.png').convert('L'))

plt.figure(dpi=200)
plt.imshow(im,cmap='gray'), plt.axis('off')
plt.show()
```

Ahora aplicaremos a esta imagen el filtro Gaussiano para valores de $\sigma$ entre 2 y 10

```python
from scipy.ndimage import filters

sigma = np.arange(2,12,2)

plt.figure(figsize=(16,10))
for i,s in enumerate(sigma,1):
    im2 = filters.gaussian_filter(im,s)
    
    plt.subplot(2,len(sigma),i), plt.imshow(im2,cmap='gray'), plt.axis('off')
    plt.title(r'$\sigma = {}$'.format(s))
    
    plt.subplot(2,len(sigma),i+len(sigma)), plt.gray()
    plt.contour(im2, origin='image'), plt.axis('off'), plt.axis('equal')

plt.show()
```

Como podemos ver, a medida de que el valor de $\sigma$ va aumentando, los detalles en la imagen cada vez se van perdiendo ya que se va poniendo más borrosa la image. Además de esto, los contornos, a medida que $\sigma$ aumenta, también se van viendo cada vez a un nivel de detalle menor ya que al estar la imagen tan borrosa es complicado ver donde están realmente los bordes, es decir, esto lo que hace es "desgastar" de cierto modo los contornos de la imagen.
